Using scikit-learn to calculate the Best Fit Line


```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#

```python
#Data for hours studied and test scores
X = np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([2,4,5,4,5])

#Create a linear regression model

model =LinearRegression().fit(X,y)

#calculate the slope(m) and intercept(c)

m = model.coef_[0]
c = model.intercept_

print(f'calculated slope(m):{m}')  # calculated slope(m): 0.6
print(f'calculated intercept(c):{c}') #calculated slope(c):2.2
```
